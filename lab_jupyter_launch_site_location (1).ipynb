```python
# Task 2: Mark the success/failed launches for each site on the map
import folium, pandas as pd
from folium.plugins import MarkerCluster

# 0) alege coloana cu numele site-ului
site_col = next(c for c in ["Launch Site","Launch_Site","LaunchSite"] if c in spacex_df.columns)

# 1) normalizează denumirile uzuale
spacex_df[site_col] = spacex_df[site_col].replace({
    "CCAFS LC-40": "CCAFS SLC 40",
    "KSC LC-39A": "KSC LC 39A",
    "VAFB SLC-4E": "VAFB SLC 4E",
})

# 2) adaugă coordonatele prin merge
site_coords = pd.DataFrame([
    {"LaunchSite":"CCAFS SLC 40","Latitude":28.561857,"Longitude":-80.577366},
    {"LaunchSite":"KSC LC 39A","Latitude":28.608058,"Longitude":-80.603955},
    {"LaunchSite":"VAFB SLC 4E","Latitude":34.632093,"Longitude":-120.610829},
])
spacex_df = spacex_df.merge(site_coords, left_on=site_col, right_on="LaunchSite", how="left")

# 3) conversie + curățare
for c in ["Latitude","Longitude"]:
    spacex_df[c] = pd.to_numeric(spacex_df[c], errors="coerce")
spacex_df["class"] = pd.to_numeric(spacex_df["class"], errors="coerce")
spacex_clean = spacex_df.dropna(subset=["Latitude","Longitude","class"]).copy()

# 4) hartă
site_map = folium.Map(
    location=[spacex_clean["Latitude"].mean(), spacex_clean["Longitude"].mean()],
    zoom_start=4
)
cluster = MarkerCluster().add_to(site_map)

for _, r in spacex_clean.iterrows():
    color = "green" if int(r["class"]) == 1 else "red"
    folium.CircleMarker(
        [r["Latitude"], r["Longitude"]], radius=5, color=color, fill=True, fill_opacity=0.7,
        popup=folium.Popup(
            f"{r[site_col]}<br>Date: {r.get('Date','N/A')}<br>"
            f"Outcome: {'Success' if int(r['class'])==1 else 'Failure'}",
            max_width=250
        )
    ).add_to(cluster)

site_map.fit_bounds([
    [spacex_clean["Latitude"].min(), spacex_clean["Longitude"].min()],
    [spacex_clean["Latitude"].max(), spacex_clean["Longitude"].max()]
])
site_map
```

```python
from folium.plugins import MarkerCluster
from math import radians, sin, cos, asin, sqrt

def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat/2)**2 + cos(radians(lat1))*cos(radians(lat2))*sin(dlon/2)**2
    return 2 * R * asin(sqrt(a))

# If you don't already have launch_sites_df, build it here
if 'launch_sites_df' not in globals():
    launch_sites_df = pd.DataFrame([
        {"Launch Site": "CCAFS LC-40", "Lat": 28.561857, "Long": -80.577366},
        {"Launch Site": "KSC LC-39A",  "Lat": 28.608058, "Long": -80.603955},
        {"Launch Site": "VAFB SLC-4E", "Lat": 34.632093, "Long": -120.610829},
    ])

POI = {
    "CCAFS LC-40": {"Coast": (28.36, -80.60), "City": (28.3200, -80.6086)},
    "KSC LC-39A":  {"Coast": (28.62, -80.57), "City": (28.3922, -80.6077)},
    "VAFB SLC-4E": {"Coast": (34.68, -120.61), "City": (34.6391, -120.4579)},
}

# Create map if it doesn't exist
if 'site_map' not in globals():
    site_map = folium.Map(location=[launch_sites_df['Lat'].mean(),
                                    launch_sites_df['Long'].mean()],
                          zoom_start=5)

# Feature groups
dist_fg = folium.FeatureGroup(name="Proximity distances").add_to(site_map)
sites_fg = folium.FeatureGroup(name="Launch sites").add_to(site_map)
pois_fg  = folium.FeatureGroup(name="Nearby POIs").add_to(site_map)

# Site markers
for _, s in launch_sites_df.iterrows():
    folium.Marker(
        [s["Lat"], s["Long"]],
        icon=folium.Icon(color="blue", icon="rocket", prefix="fa"),
        popup=f"<b>{s['Launch Site']}</b>"
    ).add_to(sites_fg)

# POI markers + distance lines
rows = []
for _, s in launch_sites_df.iterrows():
    site_name, s_lat, s_lon = s["Launch Site"], s["Lat"], s["Long"]
    if site_name not in POI:
        continue
    for label, (p_lat, p_lon) in POI[site_name].items():
        d_km = haversine_km(s_lat, s_lon, p_lat, p_lon)
        rows.append({"Launch Site": site_name, "Target": label, "Distance_km": round(d_km, 2)})
        # POI marker
        folium.CircleMarker(
            [p_lat, p_lon], radius=6, color="darkgreen", fill=True, fill_opacity=0.8,
            popup=f"{label} near {site_name}"
        ).add_to(pois_fg)
        # Line with tooltip distance
        folium.PolyLine(
            [(s_lat, s_lon), (p_lat, p_lon)], color="gray", weight=2,
            tooltip=f"{site_name} → {label}: ~{d_km:.1f} km"
        ).add_to(dist_fg)

# Fit map to all geometry
site_map.fit_bounds([
    [min(launch_sites_df["Lat"].min(), min(p[0] for d in POI.values() for p in d.values())),
     min(launch_sites_df["Long"].min(), min(p[1] for d in POI.values() for p in d.values()))],
    [max(launch_sites_df["Lat"].max(), max(p[0] for d in POI.values() for p in d.values())),
     max(launch_sites_df["Long"].max(), max(p[1] for d in POI.values() for p in d.values()))]
])

folium.LayerControl().add_to(site_map)

# Distance summary table (optional)
distances_df = pd.DataFrame(rows).sort_values(["Launch Site", "Target"])
print(distances_df)

site_map  # keep as the last line to render the map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
